```python
import pandas as pd
import json
import matplotlib.pyplot as plt
```

```python
with open('/Users/nhotin/Documents/GitHub/LegalBizAI_project/data/testset/id_cof/chunk_sz_fl_point/quaset_final.json', 'r', encoding='utf-8') as file:
    data = json.load(file)
df = pd.json_normalize(data)
```

```python
type_question_counts = df['type_question'].value_counts()
print(type_question_counts)
```

```python
plt.figure(figsize=(10, 6))
type_question_counts.plot(kind='bar', color='skyblue')
plt.title('Phân bố các loại câu hỏi trong cột "type_question"')
plt.xlabel('Loại câu hỏi')
plt.ylabel('Số lượng')
plt.xticks(rotation=0)
plt.tight_layout()

plt.show()
```
