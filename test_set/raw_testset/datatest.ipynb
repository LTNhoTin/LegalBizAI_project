```python
import matplotlib.pyplot as plt
type_q_counts = df['Type_question'].value_counts()
```

```python
# Plotting the bar chart
plt.figure(figsize=(10, 6))
type_q_counts.plot(kind='bar')
plt.title('Classification of Question Types')
plt.xlabel('type_questions')
plt.ylabel('Quantity')
plt.xticks(rotation=0)
plt.grid(axis='y')

# Show the plot
plt.show()
```
