```python
trust_lv_counts = df['Trust_lv'].value_counts()
print("Mức độ tin cậy:")
print(trust_lv_counts)
```

```python
import matplotlib.pyplot as plt
type_q_counts = df['Type_question'].value_counts()
```

```python
# Plotting the bar chart
plt.figure(figsize=(10, 6))
type_q_counts.plot(kind='bar')
plt.title('Phân loại các loại câu hỏi')
plt.xlabel('Loại câu hỏi')
plt.ylabel('Số lượng')
plt.xticks(rotation=0)
plt.grid(axis='y')

# Show the plot
plt.show()
```
